```python
from pandas import read_csv
from matplotlib import pyplot as plt
# from matplotlib import rcParams

# rcParams['font.family'] = 'serif'
# rcParams['font.serif'] = ['Times New Roman']
# rcParams['font.size'] = 18

# load the dataset
path = 'results_scale.csv'
data = read_csv(path)

data['n'] *= 77

data.head()
```

```python
# plot the results

fig, ax = plt.subplots()
ax.plot(data['n'], data['time'])
ax.set_xlabel('Количество иероглифов')
ax.set_ylabel('Среднее время, мкс')
plt.grid()

plt.savefig('research_scale.pdf', bbox_inches='tight')
plt.show()
```
